# STDP with delay learning - Replicating Nadafian et al. 2020

```python
from makeinput import make_input, plot_input
import numpy as np
import matplotlib.pyplot as plt
```

```python
nb_syn, noise_density, simtime, stimtime = 50, 2e-4, 5e3, 2e2
repetitions = int(simtime/(stimtime*10))
noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime)
plot_input(noise, stim)
print(f'SNR: {stim.shape[1]/noise.shape[1]}')
```

## Trying the STDP learning rule with no constraint

```python
class neuron(object):
    def __init__(self,nb_syn, tau_STDP_pos, tau_STDP_neg, B_pos, B_neg):
        self.delays = np.random.randint(0,tau_STDP_pos,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])*self.delays.mean()
        self.tau_STDP_pos = tau_STDP_pos
        self.tau_STDP_neg = tau_STDP_neg
        self.B_pos = B_pos
        self.B_neg = B_neg
        
    def learn(self, stream):
        mean_diff = []
        mean_delay = []
        timestamp, address = stream[1], stream[0]
        for ev in range(1,len(address)):
             #self.time_diff += (timestamp[ev]-timestamp[ev-1])
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            mean_diff.append(self.time_diff.mean())
            mean_delay.append(self.delays.mean())
            G = np.zeros([len(self.delays)])
            G[self.time_diff>self.delays] = -self.B_neg*np.exp(-(self.time_diff[self.time_diff>self.delays]-self.delays[self.time_diff>self.delays])/self.tau_STDP_neg)
            G[self.time_diff<self.delays] = self.B_pos*np.exp((self.time_diff[self.time_diff<self.delays]-self.delays[self.time_diff<self.delays])/self.tau_STDP_pos)
            self.delays += G
        return mean_diff, mean_delay
```

```python
N = neuron(nb_syn, stimtime/2, stimtime/2, .5, .5)
simtime = 1e6
repetitions = int(simtime/(stimtime*10))
noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime)
print(f'SNR: {stim.shape[1]/noise.shape[1]}')
mean_diff, mean_delay = N.learn(aer);
plt.eventplot(N.delays[:,np.newaxis]);
```

```python
plt.plot(mean_diff, label='mean delay');
plt.plot(mean_delay, label='mean ISI');
```

when replicating Nadafian et al. 2020 with simple rules and no normalization of the delays, one can observe that the STDP on synaptic delays will make them diverge. We also observe negative values -> stop condition in the article? (they make this stop condition to avoid delays to all go to zero)  

Indeed, in the original learning rule, the learning is diverging. For delays above time_diff, i.e $\Delta t_{i,j} \lt 0$, they will keep increasing because: $d_{i,j}\nearrow -> \Delta t_{i,j} \searrow$ so $p(\Delta t_{i,j} \lt 0)\nearrow$.
Same thing when delays are decreasing -> need for a stop condition

## Set a minimum value for the delay to avoid zero and negative values 

```python
# here we add d_min to set a minimum, positive value for the delays
class neuron(object):
    def __init__(self,nb_syn, tau_STDP_pos, tau_STDP_neg, B_pos, B_neg, d_min):
        self.delays = np.random.randint(d_min,tau_STDP_pos,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])*self.delays.mean()
        self.tau_STDP_pos = tau_STDP_pos
        self.tau_STDP_neg = tau_STDP_neg
        self.B_pos = B_pos
        self.B_neg = B_neg
        self.d_min = d_min
        
    def learn(self, stream):
        timestamp, address = stream[1], stream[0]
        for ev in range(1,len(address)):
            #self.time_diff += (timestamp[ev]-timestamp[ev-1])
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            G = np.zeros([len(self.delays)])
            G[self.time_diff>self.delays] = -self.B_neg*np.exp(-(self.time_diff[self.time_diff>self.delays]-self.delays[self.time_diff>self.delays])/self.tau_STDP_neg)
            G[self.time_diff<self.delays] = self.B_pos*np.exp((self.time_diff[self.time_diff<self.delays]-self.delays[self.time_diff<self.delays])/self.tau_STDP_pos)
            self.delays += G
            self.delays[self.delays<self.d_min] = self.d_min
```

```python
tau_STDP_neg = tau_STDP_pos = stimtime/2 # one condition on the article (tau_STDP>=B)
B_pos = 1e-2
B_neg = 1e-2
d_min = 1e-4

N = neuron(nb_syn, tau_STDP_pos, tau_STDP_neg, B_pos, B_neg, d_min)
for i in range(10):
    noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime)
    N.learn(aer);
    plt.eventplot(N.delays[:,np.newaxis]);
    plt.show()
```

Again we see that we learn the stimulus pattern but the delays diverge, need to normalize or change the learning rule

## Normalize delays between 0 and 1 

```python
# initialize delays between 0 and 1 and normalize between 0 and 1 at each learning step
class neuron(object):
    def __init__(self,nb_syn, tau_STDP_pos, tau_STDP_neg, B_pos, B_neg, d_min):
        self.delays = np.random.rand(nb_syn)
        self.time_diff = np.ones([nb_syn])*tau_STDP_pos
        self.tau_STDP_pos = tau_STDP_pos
        self.tau_STDP_neg = tau_STDP_neg
        self.B_pos = B_pos
        self.B_neg = B_neg
        self.d_min = d_min
        
    def learn(self, stream):
        timestamp, address = stream[1], stream[0]
        for ev in range(1,len(address)):
            #self.time_diff += (timestamp[ev]-timestamp[ev-1])
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            G = np.zeros([len(self.delays)])
            G[self.time_diff>self.delays*self.tau_STDP_neg] = -self.B_neg*np.exp(-(self.time_diff[self.time_diff>self.delays*self.tau_STDP_neg]-self.delays[self.time_diff>self.delays*self.tau_STDP_neg]*self.tau_STDP_neg)/self.tau_STDP_neg)
            G[self.time_diff<self.delays*self.tau_STDP_pos] = self.B_pos*np.exp((self.time_diff[self.time_diff<self.delays*self.tau_STDP_pos]-self.delays[self.time_diff<self.delays*self.tau_STDP_pos]*self.tau_STDP_pos)/self.tau_STDP_pos)
            self.delays += G
            self.delays[self.delays<self.d_min] = self.d_min
            self.delays /= self.delays.max()
```

```python
tau_STDP_neg = tau_STDP_pos = stimtime/2
B_pos = 1e-4
B_neg = 1e-4
d_min = 1e-4

N = neuron(nb_syn, tau_STDP_pos, tau_STDP_neg, B_pos, B_neg, d_min)
for i in range(10):
    noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime)
    N.learn(aer);
    plt.eventplot(N.delays[:,np.newaxis]*N.tau_STDP_pos);
    plt.show()
```

Not diverging but the pattern does not seem to be that stable 

## Change the learning rule for tanh instead of STDP

```python
simtime = 1e6
repetitions = int(simtime/(stimtime*10))
tau = stimtime/2
B = 1e-2
```

```python
class neuron_tanh(object):
    def __init__(self,nb_syn, tau, B):
        self.delays = np.random.randint(0,tau,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])
        self.tau = tau
        self.B = B
        
    def learn(self, stream):
        timestamp, address = stream[1], stream[0] # timestamp = aer[1] = t des spikes / address = aer[0] = i des spikes 
        print(len(timestamp))
        print(range(1,len(address)))
        for ev in range(1,len(address)):
            self.time_diff += timestamp[ev]-timestamp[ev-1]
            self.time_diff[address[ev]] = 0#(timestamp[ev]-timestamp[ev-1]) #ça ne compare le spike qu'à 1 spike ? Le spike d'avant ? 
            G = self.B*np.tanh((self.time_diff-self.delays)/self.tau) 
            self.delays += G
```

```python
tau = stimtime/2
B = 1e-2
noise_density = 2e-4
N = neuron_tanh(nb_syn, tau, B)
for i in range(10):
    print('N_epoch', i)
    noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime)
    N.learn(aer);
    plt.eventplot(N.delays[:,np.newaxis]);
    plt.show()
```

Fast convergence and stability of the kernel with this learning function. No need to normalize. No need to set a minimum delay condition. The learning will minimize the difference between time_diff and delays. It will learn the apparition delays, i.e $t_j-t_i$, of the different spikes, not the delay needed to obtain polychronous groups

# Condition for the neuron to spike
## Oscillatory behavior

```python
class neuron_tanh(object):
    def __init__(self,nb_syn, tau, B):
        self.delays = np.random.randint(0,tau,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])
        self.tau = tau
        self.B = B
        
    def learn(self, stream, T):
        timestamp, address = stream[1], stream[0]
        for ev in range(1,len(address)):
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            if timestamp[ev]//T != timestamp[ev-1]//T:
                G = self.B*np.tanh((self.time_diff-self.delays)/self.tau)
                self.delays += G
```

```python
noise_density = 2e-4

N = neuron_tanh(nb_syn, tau, B)
for i in range(100):
    noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime, cospattern)
    N.learn(aer, 1000);
plt.eventplot(N.delays[:,np.newaxis]);
plt.show()
```

## LIF with membrane threshold

```python
class neuron_tanh(object):
    def __init__(self,nb_syn, tau, B):
        self.delays = np.random.randint(0,tau,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])
        self.tau = tau
        self.tau_LIF = tau
        self.B = B
        self.nb_learn = 0
        
    def learn(self, stream, threshold):
        timestamp, address = stream[1], stream[0]
        mem_pot = []
        for ev in range(1,len(address)):
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            #membrane_potential = np.mean(np.exp(-(self.time_diff-self.delays)/self.tau_LIF))
            membrane_potential = np.mean(np.exp(-(self.time_diff[self.time_diff>self.delays]-self.delays[self.time_diff>self.delays])/self.tau_LIF))
            mem_pot.append(membrane_potential)
            if membrane_potential>threshold+(1-threshold)*np.exp(-1e5/1+.1*self.nb_learn):
                G = self.B*np.tanh((self.time_diff-self.delays)/self.tau)
                self.delays += G      
                self.nb_learn += 1
        return mem_pot
```

```python
noise_density = 2e-4
threshold = .6

tau = stimtime*2

N = neuron_tanh(nb_syn, tau, B)
for i in range(10):
    noise, stim, aer = make_input(nb_syn, noise_density, repetitions, simtime, stimtime, cospattern)
    mem_pot = N.learn(aer, threshold);
    plt.eventplot(N.delays[:,np.newaxis]);
    plt.show()
```

Idea: increase membrane potential during learning to make the spikes more and more specific to the pattern

## LIF with acumulation of spikes

```python
class neuron_tanh(object):
    def __init__(self,nb_syn, tau, B):
        self.delays = np.random.randint(0,tau,[nb_syn]).astype(float)
        self.time_diff = np.ones([nb_syn])
        self.tau = tau
        self.tau_LIF = tau
        self.B = B
        
    def learn(self, stream, threshold):
        timestamp, address = stream[1], stream[0]
        membrane_potential = np.exp(-(self.time_diff-self.delays)/self.tau_LIF).mean()
        mem_pot = []
        nb_learn = 1
        for ev in range(1,len(address)):
            self.time_diff[address[ev]] = (timestamp[ev]-timestamp[ev-1])
            membrane_potential = mem_pot*np.exp(-(self.time_diff[address[ev]]-self.delays[address[ev]])/self.tau_LIF)+1
            mem_pot.append(membrane_potential)
            if membrane_potential>threshold*(1+0.0001*nb_learn): 
                G = self.B*np.tanh((self.time_diff-self.delays)/self.tau)
                self.delays += G      
                nb_learn += 1
        return mem_pot
```
